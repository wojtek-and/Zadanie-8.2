```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
data[0].rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK', 'HIGH POSN':'MAX POZ'}, inplace=True)
```

```python
albums = data[0]
```

```python
print(f"Na liście znajduje się {albums['ARTYSTA'].nunique()} pojedynczych artystów.")
```

```python
albums['ARTYSTA'].value_counts()
```

```python
print("Najczęściej na liście pojawia się Coldplay i Take That.")
```

```python
albums.rename(str.title, axis='columns', inplace=True)
albums
```

```python
albums.drop('Max Poz', axis=1, inplace=True)
albums
```

```python
albums['Rok'].value_counts(normalize=True)
```

```python
print("Najwięcej albumów z listy wyszło w roku 1987 oraz 2000.")
```

```python
print(f"Między 1960 a 1970 rokiem włącznie wydano {len(albums[albums['Rok'] <= 1990])} albumy z listy.")
```

```python
max_value_Rok = albums['Rok'].max()
print(f"Najmłodszy album na liście został wydany w {max_value_Rok} roku.")
```

```python
earliest_albums = albums.iloc[albums.groupby('Artysta')['Rok'].idxmin()]
earliest_albums
```

```python
earliest_albums.to_csv('earliest_albums.csv', index=False)
```
